NHANES - smoothing analyses
===========================

"NHANES" is the National Health and Nutrition Examination Survey.  The website for the survey is:

http://www.cdc.gov/nchs/nhanes.htm

This notebook demonstrates how smoothing can be used to explore the relationships among health variables.

NHANES data files are distributed in SAS xport format.  This is a binary format that can be read with the xport module, available here:

https://pypi.python.org/pypi/xport/0.1.0

If you don't have access to install the package into the Python tree, just unzip the archive and put the file named `xport.py` into your working directory (you can ignore the other files in the archive).

```python
import xport
import numpy as np
import pandas as pd
```

Here is a convenience function for reading an xport file.

```python
def read_xport(fname):
    """
    Returns a DataFrame containing all data from the xport file with file name `fname`.
    """
    with xport.XportReader(fname) as reader:
        data = [row for row in reader]
    data = pd.DataFrame(data)
    data = data.set_index("SEQN")
    return data
```

The NHANES files from a common wave can be merged using the `SEQN` variable.  In the next cell, we read in several NHANES data files and merge them into a single dataset.

```python
fnames = ["BMX_G.XPT", "BPX_G.XPT", "DEMO_G.XPT"]
datasets = [read_xport(x) for x in fnames]
nhanes = pd.concat(datasets, axis=1)
print nhanes.shape
```

Here we restrict the sample to people who are at least 10 years old.  Also, some people have zero height or zero weight, so we remove those people.

```python
nhanes = nhanes[nhanes["RIDAGEYR"] >= 10]
nhanes.sort("RIDAGEYR", inplace=True)
nhanes = nhanes[nhanes["BMXHT"] > 0]
nhanes = nhanes[nhanes["BMXWT"] > 0]
```

Next we smooth three biometric variables -- BMI, height, and weight against age.  Smoothing is an analysis that estimates the conditional mean of one variable given another.  For example, in the first plot, we are estimating the mean BMI for all subjects of a given age.  The `frac` argument to the smoothing function controls the smoothness of the fitted curve.  The default is too big for large data sets so we reduce it a but here.

```python
from statsmodels.nonparametric.smoothers_lowess import lowess

for vname in "BMXBMI", "BMXHT", "BMXWT":
    fit = lowess(nhanes[vname], nhanes["RIDAGEYR"], frac=0.1)
    plt.figure()
    plt.grid(True)
    plt.plot(fit[:, 0], fit[:, 1], '-', lw=4)
    plt.xlabel("Age (years)", size=15)
    plt.ylabel(vname, size=15)
```

To get a better sense for what the smoothing procedure is doing, we can plot the smoothed curve together with a scatterplot of the data.

```python
for vname in "BMXBMI", "BMXHT", "BMXWT":
    fit = lowess(nhanes[vname], nhanes["RIDAGEYR"], frac=0.1)
    plt.figure()
    plt.grid(True)
    plt.plot(nhanes["RIDAGEYR"], nhanes[vname], 'o', color='grey', alpha=0.3)
    plt.plot(fit[:, 0], fit[:, 1], '-', lw=4)
    plt.xlabel("Age (years)", size=15)
    plt.ylabel(vname, size=15)
```

### Comparing smoothed curves among subgroups

```python
for vname in "BMXBMI", "BMXHT", "BMXWT":
    fig = plt.figure(figsize=(10, 4))
    ax = fig.add_axes([0.1, 0.1, 0.7, 0.83])
    ax.grid(True)
    for gender in 1, 2:
        ii = (nhanes["RIAGENDR"] == gender)
        y = nhanes.loc[ii, vname]
        x = nhanes.loc[ii, "RIDAGEYR"]
        fit = lowess(y, x, frac=0.2)
        ax.plot(fit[:, 0], fit[:, 1], '-', lw=4, label={1: "Male", 2: "Female"}[gender])
    ha, lb = ax.get_legend_handles_labels()
    leg = plt.figlegend(ha, lb, "center right")
    leg.draw_frame(False)
    plt.xlabel("Age (years)", size=15)
    plt.ylabel(vname, size=15)
```

### Estimating variation around the mean

As with any regression analysis, the variation around the mean can be explored through the residuals.  First we make plots of the residuals against time.  The error distributions for the BMI and weight variables are somewhat right-skewed, while the error distribution for the height variable is fairly symmetric.

The simplest approach is to estimate the level of variation as a constant that holds throughout the age range.

```python
for vname in "BMXBMI", "BMXHT", "BMXWT":
    fit = lowess(nhanes[vname], nhanes["RIDAGEYR"], frac=0.1)
    resid = nhanes[vname] - fit[:,1]
    print vname, np.sqrt(np.mean(resid**2))
    plt.figure()
    plt.grid(True)
    plt.plot(fit[:, 0], resid, 'o', alpha=0.3)
    plt.xlabel("Age (years)", size=15)
    plt.ylabel(vname, size=15)
```

A follow-up question is whether the variance is constant with respect to age.  We can assess this by regressing the absolute residuals against age.  This gives a local estimate of the median absolute deviation (MAD).  Under approximately Gaussian errors, the standard deviation is approximately 1.48 times greater than the MAD.  The results suggest a somewhat lower degree of variation around the mean for younger subjects.

```python
for vname in "BMXBMI", "BMXHT", "BMXWT":
    fit = lowess(nhanes[vname], nhanes["RIDAGEYR"], frac=0.2, it=0)
    resid = nhanes[vname] - fit[:,1]
    fit = lowess(np.abs(resid), nhanes["RIDAGEYR"], frac=0.5)
    fit[:, 1] *= 1.48
    plt.figure()
    plt.grid(True)
    plt.plot(fit[:, 0], fit[:, 1], '-', lw=4)
    plt.xlabel("Age (years)", size=15)
    plt.ylabel(vname, size=15)
    plt.ylim(ymin=0)
```
