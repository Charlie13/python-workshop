Global Historical Climatology Network data
==========================================

The Global Surface Network (GSN) is a network of weather observatories that continuously collect high quality weather data. The front page for the repository is here: 

http://www.ncdc.noaa.gov/cdo-web/datasets

The dataset includes measurements of temperature, precipitation, and many other variables at a large number of observation stations.  The data are available in annual, monthly, and daily summaries.  We will use the daily summaries here.

The data are available here:

ftp://ftp.ncdc.noaa.gov/pub/data/ghcn/daily/

See the file readme.txt on the ftp site for a discussion of the file formats.

```python
import tarfile
import numpy as np
import pandas as pd
import cStringIO as sio
from statsmodels.nonparametric.smoothers_lowess import lowess
```

The station meta-data file links station identifiers to a place name and geospatioal coordinates (longitude and latitude) for each station.  Here we create a DataFrame containing the meta-data about each of the stations.

```python
# Get the station id's that have climate data in the archive.
fname = "ghcnd_gsn.tar.gz"
tfi = tarfile.open(fname)
hcn_stations = tfi.getnames()
hcn_stations = [x.replace("ghcnd_gsn/", "") for x in hcn_stations]
hcn_stations = [x.replace(".dly", "") for x in hcn_stations]
hcn_stations = set(hcn_stations)

# Create a DataFrame containing information about the stations.
fname = "ghcnd-stations.txt"
stations = []
fid = open(fname)
for v in fid:
    if v[0:11] in hcn_stations:
        stations.append([v[0:11], v[12:20], v[21:30], v[41:71].strip()])
stations = pd.DataFrame(stations, columns=("id", "latit", "longit", "name"))
stations = stations.convert_objects(convert_numeric=True)
```

The following function allows us to look up the closest station to a given geographic point.

```python
def find_stations_loc(longit, latit):
    dx = (latit - stations.latit)**2 + (longit - stations.longit)**2
    idx = np.argmin(dx)
    return stations.iloc[idx, :]
```

This is how we can use the find_stations_loc function to find the closest station to Ann Arbor, Michigan:

```python
ids = find_stations_loc(-83, 42)
print ids
```

Here is a function that retrieves all the data from a given station, for a given climate variable, in a given year.  The data are returned as a Pandas `Series` containing the requested data.  The series' index is the date, with the year set arbitrarily to 2000 so we can overplot the series for different years.

```python
def get_station(stn, vtype, year):
    """
    Retrieve data from a given station.
    
    Parameters:
    -----------
    stn : string
        Station id
    vtype : string
        Type of measurement to return (e.g. TMIN, TMAX, PRCP)
    year : integer
        Year for which to retrieve data
    """

    # The data are stored in a tar archive.
    fname = "ghcnd_gsn.tar.gz"
    tfi = tarfile.open(fname)
    
    # Define the file layout
    colspecs = [(17, 21), (11, 15), (15, 17)]
    colspecs.extend([(22+8*k, 27+8*k) for k in range(31)])
    columns = ["Type", "Year", "Month"] + [d for d in range(1, 32)]

    # Read the fixed width file for the requested station
    fid = tfi.extractfile("ghcnd_gsn/" + stn + ".dly")
    st = "\n".join([x for x in fid])
    s = sio.StringIO(st)
    data = pd.read_fwf(s, colspecs=colspecs, header=None, na_values=9999)
    data.columns = columns
    data = data[data["Type"] == vtype]
    data = data[data["Year"] == year]
    
    # Melt the days from columns into rows
    data = pd.melt(data, id_vars=["Type", "Year", "Month"], var_name="Day").dropna()
    
    # Build a date variable but use a fake year so we overplot distinct years
    data["Date"] = data.apply(lambda x : "2000-%d-%d" % (x.Month, x.Day), axis=1)
    data["Date"] = pd.to_datetime(data["Date"], format="%Y-%m-%d")

    # Temperatures are stored in tenths of degree C
    data.value = data.value.astype(np.float64) / 10
    
    data = data.set_index("Date")
    data = data.drop(["Year", "Month", "Day", "Type"], axis=1)
    data.sort_index(inplace=True)
    data = data.value
    
    return data
```

Here we retrieve the data for two years at one station.

```python
pitt_2012 = get_station(ids.id, "TMIN", 2012)
pitt_2014 = get_station(ids.id, "TMIN", 2014)
```

Here is a function that we can use to overlay several annual data sequences.  The raw data and a smoothed curve are plotted for each series.

```python
def plot_annual(series, labels, frac=0.66):
    colors = ['orange', 'purple', 'lime']
    plt.figure(figsize=(12,8))
    ax = plt.axes([0.1, 0.1, 0.77, 0.9])
    for i, (x, lab) in enumerate(zip(series, labels)):
        lw = lowess(x.values, np.arange(len(x.values)), frac=frac)
        plt.plot(x.index, x.values, color=colors[i], alpha=0.3)
        plt.plot(x.index, lw[:, 1], label=lab, color=colors[i], lw=4)
    ha, lb = plt.gca().get_legend_handles_labels()
    leg = plt.figlegend(ha, lb, "center right")
    leg.draw_frame(False)
    plt.xlabel("Time", size=15)
    plt.ylabel("Temperature (C)", size=15)

    from matplotlib.dates import MonthLocator, DateFormatter
    months = MonthLocator()
    months_fmt = DateFormatter("%b")
    ax = plt.gca()
    ax.xaxis.set_major_locator(months)
    ax.xaxis.set_major_formatter(months_fmt)
```

Here is how the above function can be used to plot several series.

```python
plot_annual((pitt_2012, pitt_2014), ("Pitt 2012", "Pitt 2014"), frac=0.4)
```
